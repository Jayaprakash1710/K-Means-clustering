# Importing Required Libraries

```python
# basic libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#preprocessing: 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#scikit:
from sklearn.cluster import KMeans

#metrics:
from sklearn.metrics import silhouette_score

#visualisation:
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

# Reading Dataset

```python
df=pd.read_csv('E-commerce.csv')
df.head()
```

# EDA

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

**Note:** As we do not need any Personal Identification columns for clustering, we can removie the ID columns

```python
x=df.iloc[:,1:]
x.head(2)
```

# Checcking the distribution of the features

```python
plt.figure(figsize=(12,12))
for i in range(len(x.columns)):
    plt.subplot(3,2,i+1)
    sns.histplot(x[x.columns[i]]) 
```

**OBSERVATION:** Few of the features in the data doesn't follow Normal distribution, so we can normalize the data instead of standardising.

# Normalization

```python
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(x)
scaled_data=pd.DataFrame(scaled_data,columns=x.columns)
scaled_data.head(2)
```

# Hyperparameter tuning

```python
wcss=[]
k_values=[]
silhouettes_score=[]
labels=[]
for i in range(2,11):
    model=KMeans(n_clusters=i)
    model.fit(scaled_data)
    wcss.append(model.inertia_)
    labels.append(list(model.labels_))
    silhouettes_score.append(silhouette_score(scaled_data,model.labels_))
    k_values.append(i)
```

# WCSS Evaluation

```python
plt.plot(k_values,wcss,color='green',linestyle='--')
plt.scatter(k_values,wcss,color='red')
plt.xlabel('K Values')
plt.ylabel('WCSS')
plt.show()
```

# Silhouette Score Evaluation

```python
plt.plot(k_values,silhouettes_score,color='green',linestyle='--')
plt.scatter(k_values,silhouettes_score,color='red')
plt.xlabel('K Values')
plt.ylabel('Silhouette Score')
plt.show()
```

**OBSERVATION:** It is clearly visible from above plots, for the K-value of 2,3,4 we have good Silhouette score and WCSS is good for the K-value of 3 and 4.

# Clustering (K=2)

```python
k2=x.copy()
k2['Prediction']=labels[0]
k2.head(5)
```

```python
k2['Prediction'].nunique()
```

```python
plt.figure(figsize=(12,6))
plt.plot(k2.groupby('Prediction').mean().T,label=['Cluster-1','Cluster-2'])
plt.legend()
plt.show()
```

# Clustering (K=3)

```python
k3=x.copy()
k3['Prediction']=labels[1]
k3.head(5)
```

```python
k3['Prediction'].nunique()
```

```python
plt.figure(figsize=(12,6))
plt.plot(k3.groupby('Prediction').mean().T,label=['Cluster-1','Cluster-2','Cluster-3'])
plt.legend()
plt.show()
```

# Clustering (K=4)

```python
k4=x.copy()
k4['Prediction']=labels[2]
k4.head(5)
```

```python
k4['Prediction'].nunique()
```

```python
plt.figure(figsize=(12,6))
plt.plot(k4.groupby('Prediction').mean().T,label=['Cluster-1','Cluster-2','Cluster-3','Cluster-4'])
plt.legend()
plt.show()
```

**OBSERVATIONS:**
1. For the K-value 2, the model fails to predict the pattern in the data and we lose some information.
2. The K-value 3 captures the pattern in the data and the K-value 4, formation of 4th cluster doesn't provide much information about the data.

**INSIGHTS:**
1. On formation of 3 clusters, we were able to capture the pattern in the data.
2. Customers that fall under cluster-2 spend high amount and they do not expect any discount and do not visit the website frequently.
3. Customers that fall under cluster-3 make purchase whenever there is good amount of discount provided.
4. Customers who belong to cluster-1 has spent very low amount, also they do not visit the website frequently.

**RECOMMENDATIONS:**
1. Cluster-1 may spend low because they do not visit or click the websites much often. Steps can be taken to make them visit the websites frequently which would encourage them to make purchase.
2. By giving good amount of discount, cluster-3 will be encouraged to spend more.
